# Exercise 2

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
S_0=1; r=0.05; sigma=0.03; N=100; dt=1/N; 
t=np.linspace(dt,1,N)    
increments=np.random.normal(0,m.sqrt(dt),int(N))
B=np.cumsum(increments)
B=np.insert(B,0,0)
t=np.insert(t,0,0)
S=S_0*np.exp((r-(sigma**2)/2)*t+sigma*B)
plt.plot(t,S)
plt.title('Reducing the volatility and increasing the drift')
plt.xlabel('t')
plt.ylabel('S(t)')
```

##### using different values of sigma 

```python
rr=0.5; sigma1=0.03;
S=S_0*np.exp((rr-(sigma1**2)/2)*t+sigma1*B)
plt.plot(t,S)
plt.title('New values of volatility and drift')
plt.xlabel('t')
plt.ylabel('S(t)')
```
